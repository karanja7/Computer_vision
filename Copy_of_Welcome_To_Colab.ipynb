```
import pandas as pd

# First Image: Population Data Wrangling
# Data from image
population_data = {
    "region": ["Toronto", "Vancouver", "Montreal", "Calgary", "Ottawa"],
    "2006": [31612897, 2116581, 3635556, 1079310, 1133633],
    "2011": [33476688, 2313328, 3824221, 1214839, 1236324],
    "2016": [5928040, 2463431, 4098927, 1392609, 1323783]
}

# Convert to DataFrame
df_population = pd.DataFrame(population_data)

# Melt (reshape) the data to long format
df_population_long = pd.melt(df_population, id_vars=["region"], var_name="year", value_name="population")

# Display reshaped population data
print(df_population_long)

# Second Image: Car Sales Data Wrangling
# Data from image
car_data = {
    "Make + Model": ["Audi A1", "Audi A2", "Audi A3", "Audi A4", "Audi A5", "Audi A6", "Audi A7", "Audi A8"],
    "mpg": [67, 58, 48, 48, 43, 34, 31, 31],
    "sales": [29991, 65972, 89920, 90696, 56673, 47942, 36026, 12819],
    "Most Popular Colour": ["Grey", "Blue", "Red", "Black", "Grey", "White", "Black", "Black"],
    "Price($)": [23305, 24525, 27974, 34856, 44752, 53193, 54816, 75231],
    "Cost($)": [16335, 18311, 15880, 18055, 42404, 44902, 45830, 46387]
}

# Convert to DataFrame
df_car = pd.DataFrame(car_data)

# Add a Profit column
df_car["Profit($)"] = df_car["Price($)"] - df_car["Cost($)"]

# Display the updated car data
print(df_car)
```
